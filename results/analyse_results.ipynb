```python
import pandas as pd
```

```python
# Load combined results
df = pd.read_csv("gnn_results/gat/gat_results_combined.csv")
```

```python
# Find the best result by test_f1
best_result = df.loc[df["test_mae"].idxmin()]
print("Best Result:")
print(best_result)
```

```python
# Group by config_id and get average test_f1 for each configuration
avg_results = df.groupby("config_id").agg({
    "test_f1": "mean",
    "test_acc": "mean",
    "test_mae": "mean",
    "test_mse": "mean",
    "test_rmse": "mean"
}).reset_index()
```

```python
# Find the best average configuration
best_avg = avg_results.loc[avg_results["test_f1"].idxmax()]
print("\nBest Average Configuration:")
print(best_avg)
```

```python
# Display top 5 configurations by average test_f1
top_configs = avg_results.sort_values(by="test_f1", ascending=False).head(5)
print("\nTop 5 Configurations:")
print(top_configs)
```
