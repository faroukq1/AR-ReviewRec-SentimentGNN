```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "../data/row/labr_row.csv"
df = pd.read_csv(file_path)
```

```python
print("=== Dataset Info ===")
print(df.info())
```

```python
print("\n=== Dataset Shape ===")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
print("\n=== First 5 Rows ===")
print(df.head())
```

```python
print("\n=== Missing Values ===")
print(df.isnull().sum())
```

```python
print("\n=== Rating / Label Distribution ===")
print(df['rating'].value_counts())
```

```python
df['review_length'] = df['review_text'].apply(lambda x: len(str(x).split()))
print("\n=== Review Length Stats (words) ===")
print(df['review_length'].describe())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['review_length'], bins=50, kde=True)
plt.title("Distribution of Review Lengths (words)")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()
```

```python
num_users = df['user_id'].nunique()
num_books = df['book_id'].nunique()
print(f"\nNumber of unique users: {num_users}")
print(f"Number of unique books: {num_books}")
```

```python
user_review_counts = df['user_id'].value_counts()
print("\n=== Reviews per User Stats ===")
print(user_review_counts.describe())
```

```python
book_review_counts = df['book_id'].value_counts()
print("\n=== Reviews per Book Stats ===")
print(book_review_counts.describe())
```
